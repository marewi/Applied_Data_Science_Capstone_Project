```python
import pandas as pd
import requests

list1 = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
df = pd.DataFrame(list1[0])
df.head()
```

```python
postal_code_list = []

for rowIndex, row in df.iterrows(): #iterate over rows
    for columnIndex, value in row.items():
        postal_code_list.append(value[:3])

borough_list = []
for rowIndex, row in df.iterrows(): #iterate over rows
    for columnIndex, value in row.items():
        end = value.find('(')
        borough_list.append(value[3:end])
        
neighborhood_list = []
for rowIndex, row in df.iterrows(): #interate over rows
    for columnIndex, value in row.items():
        start = value.find('(')
        end = value.find(')')
        neighborhood_list.append(value[start+1:end])

data = {'postal code':postal_code_list, 'borough':borough_list, 'neighborhood':neighborhood_list}

df2 = pd.DataFrame(columns=['postal code', 'borough', 'neighborhood'], data=data)
df2.drop(df2[df2['borough'].str.contains('Not assigne')].index, inplace=True)
df2.reset_index(drop=True, inplace=True)
df2.head(20)
```

```python
df3 = df2.join(geo_spatial_dataset, lsuffix='_df2', rsuffix='_geo').drop(columns=['Postal Code'])
df3.head(10)
```

## plot locations

```python
# !pip install folium
import folium
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df3['Latitude'],df3['Longitude'],df3['borough'],df3['neighborhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
from sklearn.cluster import KMeans

k=5
toronto_clustering = df3.drop(['postal code','borough','neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
# print(kmeans.labels_)
df3.insert(0, 'Cluster Labels', kmeans.labels_)
```

## plot clusters

```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors

# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df3['Latitude'], df3['Longitude'], df3['neighborhood'], df3['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
